```python
import scipy
import numpy
import pandas
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
ds = pandas.read_csv('Data3.csv')
print(ds)
```

```python
mean = ds["Salary"].mean()
ds["Salary"].fillna(mean,inplace=True)
mean_age = ds["Age"].mean()
ds["Age"].fillna(mean_age,inplace=True)
print(ds)
```

```python
X = ds.iloc[:,:-1].values
Y = ds.iloc[:,3].values
print(X)
print(Y)
```

```python
lable_encoder_x = LabelEncoder()
X[:,0] = lable_encoder_x.fit_transform(X[:,0])
print(X)
```

```python
labelEncoder_y = LabelEncoder()
Y = labelEncoder_y.fit_transform(Y)
print(Y)
```

```python
scaler_x= StandardScaler()
scaler_xx = scaler_x.fit_transform(X)
print(scaler_xx)
```

```python
from sklearn.preprocessing import MinMaxScaler
from scipy import stats
scler = MinMaxScaler()
scaler_data_x = scler.fit_transform(X)
print(scaler_data_x)
z_score = stats.zscore(scaler_data_x)
threshold = 10
outliers = numpy.where(numpy.abs(z_score)>threshold)[0]

plt.figure(figsize=(10,6))
plt.scatter(scaler_data_x[:,1],scaler_data_x[:,2],label='Data Points')
plt.scatter(scaler_data_x[outliers,0],scaler_data_x[outliers,1],marker='o',c='red',label='outliers')
plt.title('Scatter plot with outliers')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
filterde_data = numpy.delete(scaler_data_x,outliers, axis=0)
plt.scatter(filterde_data[:,1],filterde_data[:,2],label='Data Points')
plt.title('Scatter plot with outliers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
