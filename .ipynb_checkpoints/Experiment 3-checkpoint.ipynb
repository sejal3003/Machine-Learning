```python
import pandas as pd
import scipy
import numpy
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(r"C:\Users\Sejal\Desktop\ML\Data.csv")
print(df)
```

```python
mean = df["Salary"].mean()
df["Salary"].fillna(mean,inplace=True)
mean_age = df["Age"].mean()
df["Age"].fillna(mean_age,inplace=True)
print(df)
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,3].values
print(x);
print(y);
```

```python
#Encoding categorical data
#from SK learn package pre-processing label encoder
#pip install scikit-learn
lable_encoder_x = LabelEncoder()
x[:,0] = lable_encoder_x.fit_transform(x[:,0])
print(x)
```

```python
labelEncoder_y = LabelEncoder()
y = labelEncoder_y.fit_transform(y)
print(y)
```

```python
scaler_x= StandardScaler()
scaler_xx = scaler_x.fit_transform(x)
print(scaler_xx)
```

```python
from sklearn.preprocessing import MinMaxScaler
from scipy import stats
scler = MinMaxScaler()
scaler_data_x = scler.fit_transform(x)
print(scaler_data_x)
z_score = stats.zscore(scaler_data_x)
threshold = 10
outliers = numpy.where(numpy.abs(z_score)>threshold)[0]

plt.figure(figsize=(10,6))
plt.scatter(scaler_data_x[:,1],scaler_data_x[:,2],label='Data Points')
plt.scatter(scaler_data_x[outliers,0],scaler_data_x[outliers,1],marker='o',c='red',label='outliers')
plt.title('Scatter plot with outliers')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
filterde_data = numpy.delete(scaler_data_x,outliers, axis=0)
plt.scatter(filterde_data[:,1],filterde_data[:,2],label='Data Points')
plt.title('Scatter plot with outliers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
