```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Sejal\Desktop\ML\Iris.csv")
df.tail(210)
```

```python
lb=LabelEncoder();
df[['Species']]=df[['Species']].apply(LabelEncoder().fit_transform)
```

```python
x=df.iloc[:,1:5].values
y=df.iloc[:,5:].values
print(x)
print(y)
```

```python
#sc=StandardScaler()
#x=sc.fit_transform(x)
#y=sc.fit_transform(y)
#print(x)
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,n_init='auto')
y_pred=km.fit_predict(x)
y=y.reshape(len(y))
print(y)
```

```python
plt.scatter(x[y_pred == 0,0],x[y_pred == 0,1], s =100, c='blue', label = 'Iris-setosa')# for first cluster
plt.scatter(x[y_pred == 1,0],x[y_pred == 1,1], s =100, c='green', label = 'Iris-capsule')# for first cluster
plt.scatter(x[y_pred == 2,0],x[y_pred == 2,1], s =100, c='red', label = 'Iris-virginica')# for first cluster
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100,c='yellow',label='centroids')
plt.title('Clusters')
plt.legend()
plt.show()
```
