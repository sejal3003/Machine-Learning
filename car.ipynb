```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(r"C:\Users\Sejal\Desktop\ML\car_price_prediction.csv")

df.head()
```

```python
df['Mileage']=df['Mileage'].str.replace("km","")
df['Doors']=df['Doors'].str.replace("May","")
df['Engine volume']=df['Engine volume'].str.replace("Turbo","")
df['Levy']=df['Levy'].replace("-",np.NaN)
```

```python
df[['Manufacturer','Category','Model','Fuel type','Gear box type','Color','Wheel','Drive wheels','Leather interior','Doors']]=df[['Manufacturer','Category','Model','Fuel type','Gear box type','Color','Wheel','Drive wheels','Leather interior','Doors']].apply(LabelEncoder().fit_transform)
df.head(10)
```

```python
from sklearn.impute import SimpleImputer
im_mean=SimpleImputer(missing_values=np.NaN,strategy='mean')
df[['Levy']]=im_mean.fit_transform(df[['Levy']])
df.head(15)
```

```python
x=df.iloc[:,3:].values
y=df.iloc[:,1:2].values
print(x);
print(y);
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=2)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
x=sc.fit_transform(x)
y=sc.fit_transform(y)
print(x)
y
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
c=KNeighborsClassifier(n_neighbors=3,metric='minkowski',p=1)
c.fit(x_train,y_train)

```

```python
y_pred=c.predict(x_test)
print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
ac=accuracy_score(y_test,y_pred)
#print(classification_report(y_test,y_pred))
print(cm)
print(ac)
```

```python
from sklearn import metrics
m=metrics.ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[False,True])
m.plot()
m.show()
```
